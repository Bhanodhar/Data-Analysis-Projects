##### Find most famous restaurants chains(like Franchise(Rastaurants having more than one branch) in Bangalore.

```python
# Group by 'name' and count the number of branches
restaurant_chains = zr.groupby('name').size().reset_index(name='branch_count')

# Filter for restaurants with more than one branch
franchises = restaurant_chains[restaurant_chains['branch_count'] > 1]

# Sort the data to find the most famous chains
franchises = franchises.sort_values(by='branch_count', ascending=False)

# Display the top 10 most famous restaurant chains
print(franchises.head(10))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Visualize the top 10 restaurant chains
top_franchises = franchises.head(10)

plt.figure(figsize=(14, 8))
sns.barplot(data=top_franchises, x='branch_count', y='name', palette='viridis')
plt.title('Top 10 Most Famous Restaurant Chains in Bangalore')
plt.xlabel('Number of Branches')
plt.ylabel('Restaurant Chain')
plt.show()
```

#### Find how many voters gives rating for each 'type' and aggregate rating of that 'type

```python
# Convert 'votes' to numeric
zr['votes'] = pd.to_numeric(zr['votes'], errors='coerce')

# Convert 'rating' to numeric if not already done
zr['rating'] = pd.to_numeric(zr['rating'], errors='coerce')

# Group by 'type' and calculate the number of voters and aggregate rating
type_aggregation = zr.groupby('type').agg(
    total_voters=('votes', 'sum'),
    aggregate_rating=('rating', 'sum')
).reset_index()

# Display the result
print(type_aggregation)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Plot total voters for each type
plt.figure(figsize=(14, 8))
sns.barplot(data=type_aggregation, x='total_voters', y='type', palette='viridis')
plt.title('Total Voters for Each Type of Restaurant')
plt.xlabel('Total Voters')
plt.ylabel('Type of Restaurant')
plt.show()
```

```python
# Plot aggregate ratings for each type
plt.figure(figsize=(14, 8))
sns.barplot(data=type_aggregation, x='aggregate_rating', y='type', palette='viridis')
plt.title('Aggregate Ratings for Each Type of Restaurant')
plt.xlabel('Aggregate Rating')
plt.ylabel('Type of Restaurant')
plt.show()
```
